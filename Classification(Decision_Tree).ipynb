# Load Data 

```python
from sklearn import datasets

iris_data = datasets.load_iris() #วิธีการโหลด data ที่ต้องการ 
```

```python
iris_data
```

#### ความหมาย ของข้อมูลที่นำมาใช้
sepal length ความยาวของกลีบเลี้ยง

sepal width ความกว้างของกลีบเลี้ยง

petal length ความยาวของกลีบดอก

petal width  ความกว้างของกลีบดอก

target = y

```python
X = iris_data['data'] 
Y = iris_data['target']
```

# Train Model  

# import libaly ที่เราจำนำมา train

```python
from sklearn.tree import DecisionTreeClassifier
```

# define

```python
mytree = DecisionTreeClassifier(max_depth=2) # max_depth คือการปรับแต่ง ตันไม้ 2 ชั้น
```

# train

```python
mytree.fit(X,Y) # fit() คือการใช้คำสั่ง fitmodel ใส่data ที่เราได้กำหนดไว้
```

# plot tree

```python
from sklearn.tree import plot_tree
plot_tree(mytree) # ทำการ พลอต ด้วยคำสั่ง Plot_tree()
```
