# K-Nearest Neighbour

  import

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=6)
X_traintrain, X_val, y_traintrain, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=6)
```

```python
neigh1_CKD = KNeighborsClassifier(n_neighbors=1) #เชื่อใกล้สุดแค่คนเดียว
```

```python
neigh1_CKD.fit(X_traintrain,y_traintrain)
```

```python
Y_predictKNN = neigh1_CKD.predict(X_test)
Y_predictKNN  #ผลการทำนาย
```

```python
neigh1_CKD2 = KNeighborsClassifier(n_neighbors=3) #เชื่อใกล้สุด 3 คน 
neigh1_CKD2.fit(X_traintrain,y_traintrain)
```

```python
Y_predictKNN2 = neigh1_CKD2.predict(X_test)
Y_predictKNN2  #ผลการทำนาย
```

```python
from sklearn.metrics import accuracy_score
```

```python
knn1_result = neigh1_CKD.predict(X_val) #ตรวจสอบความถูกต้องของModel ที่1 เชื่อคนใกล้ 1 คน
accuracy_score(y_val,knn1_result)*100
```

```python
knn1_result2 = neigh1_CKD2.predict(X_val)  #ตรวจสอบความถูกต้องของModel ที่2 เชื่อคนใกล้ 3 คน
accuracy_score(y_val,knn1_result2)*100
```
