```python
import pythainlp
from pythainlp import word_tokenize
# from pythainlp.corpus import stopwords
# from nltk.corpus import stopwords
# import stop_words #as stopwords
from pythainlp.corpus import wordnet
from nltk.stem.porter import PorterStemmer
from nltk.corpus import words
from stop_words import get_stop_words
```

```python
from pythainlp.corpus.common import thai_stopwords
```

```python
thai_stopwords = list(thai_stopwords())
thai_stopwords
```

```python
from pythainlp import word_tokenize
def text_process(lyrics):
    final = "".join(u for u in lyrics if u not in ("?", ".", ";", ":", "!", '"', "ๆ", "ฯ"))
    final = word_tokenize(final)
    final = " ".join(word for word in final)
    final = " ".join(word for word in final.split() 
                     if word.lower not in thai_stopwords)
    return final

data['lyrics_tokens'] = data['lyrics'].apply(text_process)
data
```

#Split ข้อมูลเป็น Train (80%) Test (20%)

```python
from sklearn.model_selection import train_test_split
X = data[['lyrics_tokens']]
y = data['tarket']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

#Word Vectorizer และ Bag-of-Words (BoW)

การจะสร้างแบบจำลองวิเคราะห์จากข้อความ จำเป็นจะต้องแปลงข้อมูลให้อยู่ในรูปแบบที่มีจำนวนของตัวแปรที่ทำกัน โดยเริ่มต้นเราจะใช้ Count Vectorizer ของ sklearn มาช่วยในการจัดการดึงคำทั้งหมดออกมาจากข้อความ และจัดเก็บในรูปแบบ Vector (จะคล้าย ๆ กับลักษณะของพจนานุกรมที่มีการระบุตัวเลข index ของแต่ละคำด้วย)

```python
from sklearn.feature_extraction.text import CountVectorizer
cvec = CountVectorizer(analyzer=lambda x:x.split(' '))
cvec.fit_transform(X_train['lyrics_tokens'])
cvec.vocabulary_
```

```python
train_bow = cvec.transform(X_train['lyrics_tokens'])
pd.DataFrame(train_bow.toarray(), columns=cvec.get_feature_names(), index=X_train['lyrics_tokens'])
```

#สร้างแบบจำลอง Logistic Regression เพื่อจำแนกเพลงที่ใช้บำบัด

ทำการสร้างแบบจำลอง Logistic Regression เพื่อจำแนก (classify) ข้อความว่าเป็น บำบัดได้ ไม่แน่ใจ และ บำบัดไม่ได้ โดยให้ตัวแปรต้นเป็น BoW ที่สร้างจากข้อความสำหรับการฝึกฝน และตัวแปรตามคือ y_train หรือก็คือคอลัมน์ Tarket ที่แบ่งไว้สำหรับการฝึกฝน

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_bow, y_train)
```

#ทดสอบแบบจำลอง


```python
from sklearn.metrics import confusion_matrix,classification_report
test_bow = cvec.transform(X_test['lyrics_tokens'])
test_predictions = lr.predict(test_bow)
print(classification_report(test_predictions, y_test))
```
